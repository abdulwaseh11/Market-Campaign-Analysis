```python
for feature in ['ID', 'Education', 'Marital_Status', 'Income', 'Kidhome', 'Teenhome',
       'Dt_Customer', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
       'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
       'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
       'NumStorePurchases', 'NumWebVisitsMonth', 'AcceptedCmp3',
       'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2',
       'Response', 'Complain', 'Country', 'Age']:
    plt.figure(figsize=(10, 6))
    
    if df[feature].dtype in ['int64', 'float64']:  
        sns.boxplot(x=df[feature])
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Value')
    else:  
        sns.countplot(x=df[feature])
        plt.title(f'Count of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Count')
    
    plt.show()
```

Section 03: Data Visualization

•	Which marketing campaign is most successful?

•	Display the total amount spent by a customer in each product category

```python
product_categories = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df_spending = df[['ID'] + product_categories].copy()
```

```python
average_spending = df[product_categories].mean()
average_spending_df = pd.DataFrame(average_spending).reset_index()
average_spending_df.columns = ['Product Category', 'Average Amount Spent']
plt.figure(figsize=(12, 6))
sns.barplot(x='Product Category', y='Average Amount Spent', data=average_spending_df)
plt.title('Average Total Spending by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Amount Spent')
plt.xticks(rotation=45)
plt.show()
```

From this graph, we can conclude that most spending is done on Wines and Meat Products

While the other products have almost equal spending amount.

•	What is the average spending in each age group?

```python
bins = [0, 18, 30, 40, 50, 60, 100]  
labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+'] 
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
product_categories = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Spending'] = df[product_categories].sum(axis=1)
average_spending_by_age = df.groupby('Age_Group')['Total_Spending'].mean().reset_index()
average_spending_by_age.columns = ['Age_Group', 'Average_Spending']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Age_Group', y='Average_Spending', data=average_spending_by_age)
plt.title('Average Total Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Total Spending')
plt.xticks(rotation=45)
plt.show()
```

From this graph, we can conclude that 70+ age category has the most average spending amount

while 50-59 age category has the lowest average spending amount.
